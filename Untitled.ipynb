```python
network_1 = json.load(open('network_1.json'))
network_2 = json.load(open('network_2.json'))
```

```python
json.dump(network_1 | network_2, open('network.json', 'w'))
```

```python
clients = pd.DataFrame({
    x: dict(Counter(network_1.get(x, {})) + Counter(network_2.get(x, {})))
    for x in set(network_1.keys()) | set(network_2.keys())
}).transpose().fillna(0)
```

```python
clients.index = clients.index.map(lambda x: datetime.strptime(x, '%Y%m%d%H%M%S'))
```

```python
aliases = {
    'openethereum': [
        'openethereum',
        'parity',
        'parity-ethereum'
    ],
    'geth': [
        'geth'
    ],
    'erigon': [
        'erigon'
    ],
    'nethermind': [
        'nethermind'
    ],
    'reth': [
        'reth'
    ]
}
```

```python
for key, values in aliases.items():
    alias = clients[values].sum(axis=1)
    clients = clients.drop(columns=values)
    clients[key] = alias
```

```python
known_clients = set(aliases.keys()) | {'akula', 'besu'}
top_clients = set(clients.sum().sort_values(ascending=False)[:10].index)
index = clients.sum()[
    list(known_clients | top_clients)
].sort_values(ascending=False).index
```

```python
index = ['geth', 'openethereum', 'erigon', 'nethermind', 'besu', 'reth']
```

```python
index_percentage = clients.divide(clients.sum(axis=1), axis=0)[index]
```

```python
other = clients.divide(clients.sum(axis=1), axis=0).drop(index, axis=1).sum(axis=1)
```

```python
fig = index_percentage.plot(kind='area')
fig.update_xaxes(title='date')
fig.update_yaxes(title='nodes', tickformat=".1%", range=[0, 1])
fig.add_trace(go.Scatter(x=other.index, y=other.values, stackgroup='1', marker_color='#d4d4d4', name='other'))
fig.layout.legend.title = 'client'
fig.layout.legend
fig.show()
print(chart_studio.plotly.plot(fig, auto_open=False))
```
